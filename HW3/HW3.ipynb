```python
import numpy
from google.colab.patches import cv2_imshow
from PIL import Image
```

```python
def median_filter(data, filter_size):
    data_final = []
    data_final = numpy.zeros((len(data),len(data[0])))
    #your code
    data_pad = []
    fs = filter_size // 2
    # the number of rows in data
    for i in range(len(data)):
        # the number of columns in data
        for j in range(len(data[0])):
            #It checks that if the image size is not divisible by the desired filter, it puts zero in the array instead
            for z in range(filter_size):
                if i + z - fs < 0 or i + z - fs > len(data) - 1:
                    for c in range(filter_size):
                        data_pad.append(0)
                else:
                    if j + z - fs < 0 or j + fs > len(data[0]) - 1:
                        data_pad.append(0)
                    else:
                        # Adds value of the image's pixels to the array
                        for k in range(filter_size):
                            data_pad.append(data[i + z - fs][j + k - fs])
            # Sort the array
            data_pad.sort()
            # find the median in array and put it in the central pixel
            data_final[i][j] = data_pad[len(data_pad) // 2]
            # Empty the array for the next column or row
            data_pad = []



    return data_final
```

```python
#do not change this cell
img = Image.open("/content/noisyimg.png").convert(
        "L")
arr = numpy.array(img)
removed_noise = median_filter(arr, 1) 
cv2_imshow(removed_noise)
```
